```python
import random
import matplotlib.pyplot as plt
def generateRandomNumbers(number,start,end):
    random_numbers=[]
    for i in range(number):
         random_numbers.append(random.randint(start,end))
    return random_numbers
def main():
   number,start,end=100,1,100
   random_numbers=generateRandomNumbers(number,start,end)
   print('Generated Random Numbers:',random_numbers)
   plt.hist(random_numbers,bins=10,color='pink',edgecolor='red')
   plt.xlabel('Generated Random Numbers Range 1 to 100 Including')
   plt.ylabel('Frequency Count')
   plt.title('Generated Random Numbers Range 1 to 100 Including Vs Frequency Count')
   plt.show()
output=main()
print(output)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets using a 70-30 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the k-NN classifier
k = 3  # Number of neighbors
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of k-NN model: {accuracy:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_points(dimensions, num_points=100):
    return np.random.rand(num_points, dimensions)

def calculate_min_distances(dimensions, num_points=100):
    min_distances = []
    for dim in dimensions:
        points = generate_points(dim, num_points)
        min_distance = np.inf
        for i in range(num_points):
            for j in range(i + 1, num_points):
                distance = np.linalg.norm(points[i] - points[j])
                if distance < min_distance:
                    min_distance = distance
        min_distances.append(min_distance)
    return min_distances

# Define the range of dimensions to explore
dimensions = range(1, 21)

# Calculate minimum distances for each dimension
min_distances = calculate_min_distances(dimensions)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(dimensions, min_distances, marker='o')
plt.xlabel('Number of Dimensions')
plt.ylabel('Minimum Distance')
plt.title('Minimum Distance Between Points vs. Number of Dimensions')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
weight = np.array([50, 60, 70, 80, 90, 100])
height = np.array([150, 155, 160, 165, 170, 175])

# Step 1: Least-Squares Estimation
# Calculate the means
mean_weight = np.mean(weight)
mean_height = np.mean(height)

# Calculate the coefficients
numerator = np.sum((weight - mean_weight) * (height - mean_height))
denominator = np.sum((weight - mean_weight) ** 2)
slope = numerator / denominator
intercept = mean_height - slope * mean_weight

# Step 2: Calculate R-squared value
predicted_height = slope * weight + intercept
ss_total = np.sum((height - mean_height) ** 2)
ss_residual = np.sum((height - predicted_height) ** 2)
r_squared = 1 - (ss_residual / ss_total)

# Step 3: Plot scatter plot and regression line
plt.figure(figsize=(10, 6))
plt.scatter(weight, height, color='blue', label='Data points')
plt.plot(weight, predicted_height, color='red', label='Regression line')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Height vs Weight Regression')
plt.legend()
plt.grid(True)
plt.show()

# Output the results
print(f"Estimated slope (m): {slope:.2f}")
print(f"Estimated intercept (b): {intercept:.2f}")
print(f"R-squared value: {r_squared:.2f}")
```
