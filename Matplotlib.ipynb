```python
#from a pandas dataframe
df=pd.read_csv('/content/sharma-kohli.csv')
```

```python
plt.plot(df['index'],df['V Kohli'])
```

```python
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
```

```python
#Label titile
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
```

```python
#colors
#Label titile
plt.plot(df['index'],df['V Kohli'],color='green')
plt.plot(df['index'],df['RG Sharma'],color='black')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
```

```python
#what if we dont want solid lines
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed')
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle='dashed')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
```

```python
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dotted')
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle='dotted')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
```

```python
#linewidth
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed',linewidth=3)
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle='dashed',linewidth=5)
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
```

```python
#if we want to mark the points as something
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed',marker='D',label="kohli")
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle='dashed',label="sharma")
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.legend()
```

```python
#limiting axes
price=[48000,56000,65000,72000,71000,89000]
years=[2015,2016,2017,2018,2019,2020]
plt.plot(years,price)
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.xlim(2015,2020)
plt.ylim(40000,90000)
```

```python
#grid
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed',marker='D',label="kohli")
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle='dashed',label="sharma")
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.legend()
plt.grid()
```

```python
#show
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed',marker='D',label="kohli")
plt.plot(df['index'],df['RG Sharma'],color='black',linestyle='dashed',label="sharma")
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.legend()
plt.grid()
plt.show()
```
